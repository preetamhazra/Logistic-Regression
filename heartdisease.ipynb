```python
import tensorflow as tf
import keras
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import mean_squared_error as skmse
from sklearn.preprocessing import MinMaxScaler, StandardScaler, RobustScaler
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Scaling function #
def Scaling(x,choice):
    if choice==1:
        scaler=MinMaxScaler()
    elif choice==2:
        scaler=StandardScaler()
    elif choice==3:
        scaler=RobustScaler()
    else:
        return (print('bad input!!'))
    mod=scaler.fit(x)
    x=mod.transform(x)
    return x
```

```python
def mmm(n,choice):
    print('initial number of NaNs=',hd.iloc[:,n].isna().sum())
    if choice=='mean':
        hd.iloc[:,n]=hd.iloc[:,n].fillna(hd.iloc[:,n].mean())
        print('updated number of NaNs=',hd.iloc[:,n].isna().sum())
        #return hd.iloc[:,n]
    elif choice=='median':
        hd.iloc[:,n]=hd.iloc[:,n].fillna(hd.iloc[:,n].median())
        print('updated number of NaNs=',hd.iloc[:,n].isna().sum())
        #return hd.iloc[:,n]
    elif choice=='mode':
        hd.iloc[:,n]=hd.iloc[:,n].fillna(hd.iloc[:,n].mode().values[0])
        print('updated number of NaNs=',hd.iloc[:,n].isna().sum())
        #return hd.iloc[:,n]
    else:
        return print('wrong input; try again!!')
```
